```python
# Importing Libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import yfinance as yf

# This shows an example of the "fivethirtyeight" styling, which tries to 
# replicate the styles from FiveThirtyEight.com.
plt.style.use("fivethirtyeight")
```

```python
# Importing and storing the Data
USD = yf.download("USD", "2019-07-14", "2020-07-15")

# https://pypi.org/project/yfinance/
#Shows the data
USD
```

```python
# visualizing the Data
plt.figure(figsize=(12.5, 4.5))
plt.plot(USD, label = "USD")
plt.title("Yahoo Finances Adj Close Price")
plt.xlabel("July 2019 - July 2020")
plt.ylabel("Adj. Close Price USD ($)")
plt.legend(loc = "upper left")
plt.show()
```
